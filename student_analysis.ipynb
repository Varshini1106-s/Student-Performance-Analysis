# Data Cleaning

The dataset was checked for missing values and duplicate records. No missing values were found, and duplicate records were removed if present.

# Data Analysis

In this section, key insights are extracted from the dataset, including average final grades, student performance, study time correlation, and gender-wise comparison.

```python
print("Average Final Grade:", df['G3'].mean())
```

```python
print("Students scoring above 15:", len(df[df['G3'] > 15]))
```

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and final grade:", correlation)
```

```python
gender_avg = df.groupby('sex')['G3'].mean()
print(gender_avg)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color ='skyblue',edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

gender_avg.index = ['Female', 'Male']

plt.figure(figsize=(6,4))
gender_avg.plot(
    kind='bar',
    color=['pink', 'blue']
)

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
